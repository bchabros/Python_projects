## Inference class to infer the mask R-CNN Model

```python
class InferenceConfig(coco.CocoConfig):
    # Setting batch size equal to 1 since we will be running inference on
    # one image at a time. Batch size = GPU_COUNT * IMAGE_PER_GPU
    GPU_COUNT = 1 
    IMAGES_PER_GPU = 1
    
config =InferenceConfig()
config.display()
```

## Loading the weights

```python
cd '/Users/Chabi/Documents/Python_projects/Image_Segmenttion/Mask_RCNN/'
```

```python
# Create model objects in inference model mode
model = modellib.MaskRCNN(mode='inference',
                          model_dir='mask_rcnn_coco.hy',
                          config=config)

# Load weights trained on MS-COCO
model.load_weights('mask_rcnn_coco.h5', 
                   by_name=True)
```

## Loading an Image to Test the Model

```python
image = skimage.io.imread('images/4410436637_7b0ca36ee7_z.jpg')

# original image
plt.figure(figsize=(12,10))

skimage.io.imshow(image);
```

## Sending Image to Model to Generate Predict

```python
# Run detection
results = model.detect([image], verbose=1)
```

## Masking the Results to image

```python
# Visualize results
r = results[0]
visualize.display_instances(image,
                            r['rois'],
                            r['masks'],
                            r['class_ids'],
                            r['scores'])
```
